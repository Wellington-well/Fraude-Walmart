## Investigação por data

```python
# Conversão da coluna 'date' para datetime
df_geral['date'] = pd.to_datetime(df_geral['date'])

# Criando colunas auxiliares: mês e semana do mês
df_geral['month'] = df_geral['date'].dt.to_period('M').astype(str)
df_geral['day'] = df_geral['date'].dt.day

# Função para categorizar a semana do mês
def week_of_month(day):
    if day <= 7:
        return '1st_week'
    elif day <= 14:
        return '2nd_week'
    elif day <= 21:
        return '3rd_week'
    else:
        return '4th_week'

df_geral['week_of_month'] = df_geral['day'].apply(week_of_month)

# Agrupamentos
items_missing_by_month = df_geral.groupby('month')['items_missing'].sum().reset_index()
items_missing_by_week = df_geral.groupby(['month', 'week_of_month'])['items_missing'].sum().reset_index()

# Estilo dos gráficos
sns.set_theme(style="whitegrid")

# Gráfico 1: Items Missing por Mês
plt.figure(figsize=(10, 10))
sns.barplot(data=items_missing_by_month, x='month', y='items_missing', palette='Blues_d')
plt.title('Items Missing por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Items Missing')
plt.show()

# Gráfico 2: Items Missing por Semana para cada Mês em subplots
unique_months = items_missing_by_week['month'].unique()
n_months = len(unique_months)

# Define o layout dos subplots (automático)
n_cols = 4
n_rows = (n_months + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 4 * n_rows), constrained_layout=True)

# Caso tenha apenas 1 linha ou 1 coluna, transformar axes em array plano
axes = axes.flatten()

for i, month in enumerate(unique_months):
    ax = axes[i]
    data_month = items_missing_by_week[items_missing_by_week['month'] == month]
    
    sns.barplot(data=data_month, x='week_of_month', y='items_missing', palette='pastel', ax=ax)
    
    ax.set_title(f'Items Missing por Semana - {month}')
    ax.set_xlabel('Semana do Mês')
    ax.set_ylabel('Quantidade')

# Se houver mais subplots que meses, remove os vazios
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle('Items Missing por Semana - Separado por Mês', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupamento por região
region_missing = df_geral.groupby('region')['items_missing'].sum().sort_values(ascending=False)

# Agrupamento por hora de entrega (ordenado do maior para o menor)
hour_missing = df_geral.groupby('hour_delivery')['items_missing'].sum().sort_values(ascending=False)

# Agrupamento por dia
day_missing = df_geral.groupby('day')['items_missing'].sum().sort_index()

# Plotando os gráficos
plt.figure(figsize=(18, 5))

# Gráfico por região
plt.subplot(1, 3, 1)
sns.barplot(x=region_missing.index, y=region_missing.values, palette="Blues_d")
plt.title('Items Missing por Região')
plt.xlabel('Região')
plt.ylabel('Total de Items Missing')
plt.xticks(rotation=45)

# Gráfico por hora de entrega (ordenado)
plt.subplot(1, 3, 2)
sns.barplot(x=hour_missing.index, y=hour_missing.values, palette="Greens_d")
plt.title('Items Missing por Hora de Entrega')
plt.xlabel('Hora')
plt.ylabel('Total de Items Missing')

# Gráfico por dia
plt.subplot(1, 3, 3)
sns.barplot(x=day_missing.index, y=day_missing.values, palette="Oranges_d")
plt.title('Items Missing por Dia')
plt.xlabel('Dia')
plt.ylabel('Total de Items Missing')

plt.tight_layout()
plt.show()
```

```python
def classificar_periodo(hora):
    if 0 <= hora <= 4:
        return 'Madrugada'
    elif 5 <= hora <= 11:
        return 'Manhã'
    elif 12 <= hora <= 14:
        return 'Almoço'
    elif 15 <= hora <= 17:
        return 'Tarde'
    elif 18 <= hora <= 23:
        return 'Noite'
    else:
        return 'Desconhecido'

# Converte a coluna para inteiro, se necessário
df_geral['hour_delivery'] = pd.to_numeric(df_geral['hour_delivery'], errors='coerce')

# Aplica a função
df_geral['periodo_dia'] = df_geral['hour_delivery'].apply(classificar_periodo)

# Etapa 2: Agrupar por período do dia
periodo_missing = df_geral.groupby('periodo_dia')['items_missing'].sum()

# Etapa 3: Ordenar os períodos manualmente
ordem_periodos = ['Madrugada', 'Manhã', 'Almoço', 'Tarde', 'Noite']
periodo_missing = periodo_missing.reindex(ordem_periodos)

# Etapa 4: Plotar o gráfico
plt.figure(figsize=(8, 5))
sns.barplot(x=periodo_missing.index, y=periodo_missing.values, palette="Purples_d")
plt.title('Items Missing por Período do Dia')
plt.xlabel('Período do Dia')
plt.ylabel('Total de Items Missing')
plt.tight_layout()
plt.show()
```

```python
# Etapa 2: Criar coluna do período do dia
def classificar_periodo(hora):
    if 0 <= hora <= 4:
        return 'Madrugada'
    elif 5 <= hora <= 11:
        return 'Manhã'
    elif 12 <= hora <= 14:
        return 'Almoço'
    elif 15 <= hora <= 17:
        return 'Tarde'
    elif 18 <= hora <= 23:
        return 'Noite'
    else:
        return 'Desconhecido'

df_geral['periodo_dia'] = df_geral['hour_delivery'].apply(classificar_periodo)

# Etapa 3: Agrupar por mês e período
periodo_mes = (
    df_geral.groupby(['month', 'periodo_dia'])['items_missing']
    .sum()
    .reset_index()
)

# Etapa 4: Garantir ordem dos períodos do dia
ordem_periodos = ['Madrugada', 'Manhã', 'Almoço', 'Tarde', 'Noite']
periodo_mes['periodo_dia'] = pd.Categorical(periodo_mes['periodo_dia'], categories=ordem_periodos, ordered=True)

# Etapa 5: Plotar gráficos por mês
meses_unicos = sorted(periodo_mes['month'].unique())
num_meses = len(meses_unicos)

fig, axes = plt.subplots(nrows=num_meses, figsize=(10, 4 * num_meses), sharex=True)

if num_meses == 1:
    axes = [axes]  # Se houver apenas um mês

for i, mes in enumerate(meses_unicos):
    dados_mes = periodo_mes[periodo_mes['month'] == mes]
    sns.barplot(
        ax=axes[i],
        x='periodo_dia',
        y='items_missing',
        data=dados_mes,
        palette='Purples_d'
    )
    axes[i].set_title(f'Items Missing por Período do Dia - Mês {int(mes)}')
    axes[i].set_ylabel('Items Missing')
    axes[i].set_xlabel('Período do Dia')

plt.tight_layout()
plt.show()
```

```python
# Tabela dinâmica com soma de items_missing
heatmap_data = df_geral.pivot_table(
    index='region',
    columns='periodo_dia',
    values='items_missing',
    aggfunc='sum',
    fill_value=0
)

# Ordenar regiões por total geral de items_missing
heatmap_data['total'] = heatmap_data.sum(axis=1)
heatmap_data = heatmap_data.sort_values('total', ascending=False).drop(columns='total')

# Plotar o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='Purples')
plt.title('Items Missing por Região e Período do Dia')
plt.xlabel('Período do Dia')
plt.ylabel('Região')
plt.tight_layout()
plt.show()
```

```python
# Criar faixas de valor do pedido
bins = [0, 100, 200, 500, 1000, 2000, float('inf')]
labels = ['até 100', '101-200', '201-500', '501-1000', '1001-2000', '2000+']
df_geral['faixa_valor_pedido'] = pd.cut(df_geral['order_amount'], bins=bins, labels=labels)

# Agrupar por faixa e região
agrupado = df_geral.groupby(['faixa_valor_pedido', 'region'])['items_missing'].sum().reset_index()

# Pivotar para heatmap
pivot = agrupado.pivot(index='faixa_valor_pedido', columns='region', values='items_missing').fillna(0)

# Plotar o heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Items Missing por Faixa de Valor do Pedido e Região')
plt.xlabel('Região')
plt.ylabel('Faixa de Valor do Pedido')
plt.tight_layout()
plt.show()
```
